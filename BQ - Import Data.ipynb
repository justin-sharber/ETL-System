## Detect Files

```python
#- read prior file list
listA = read_file_list()
listA.shape
```

```python
#- scan current files
listB = glob_files().str.lower()
listB.shape
```

```python
## new files
idx = listB.isin(listA)
newFiles = listB[~idx]
newFiles.shape
```

```python
#- folders rundown
#newFiles.str.split('/').str[3].value_counts()#.sort_index()
```

## Master Loop

```python
#- load master list
masterList = loadMasterList()
masterList.shape
```

```python
## MASTER LOOP
#- loop over data sources in master list.
for index, row in masterList.iterrows():
    info = row
    source = info.DataSource.upper()

    #- fix case
    path = info.Folder.lower()
    files = [file.lower() for file in newFiles if path in file]
    
    #- loop over files that match.
    if len(files) > 0:
        print(); print('>>', source)

    for fp in files:    
        fn = fp.split('/')[-1]
        print(fn)

        prepFile(info, fp)


print(); print('--Complete--')        
```
